```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import time

def compute_result(size):
    count=1
    temp=1
    for row in range(1,size + 1):
        for column in range(1, size + 1):
            count+=1
            temp=row + column
    return count
input_sizes=np.arange(1, 1001)
execution_times=[]
for size in input_sizes:
    start_time = time.time()
    compute_result(size)
    end_time = time.time()
    execution_times.append(end_time - start_time)
def quadratic_fit(x_values, coeff_a, coeff_b, coeff_c):
    return coeff_a * x_values**2 + coeff_b * x_values + coeff_c
def cubic_model(x_values, coeff_d):
    return coeff_d * x_values**3
def linear_model(x_values, coeff_e):
    return coeff_e * x_values
quad_coeffs, _ = curve_fit(quadratic_fit, input_sizes, execution_times)
cubic_coeffs, _ = curve_fit(cubic_model, input_sizes, execution_times)
linear_coeffs, _ = curve_fit(linear_model, input_sizes, execution_times)
plt.figure(figsize=(12, 7))
plt.plot(input_sizes, execution_times, label='Measured Time', color='blue')
plt.plot(input_sizes, quadratic_fit(input_sizes, *quad_coeffs), color='red', label='Quadratic Fit')
plt.plot(input_sizes, cubic_model(input_sizes, *cubic_coeffs), '--', linestyle='--', color='green', label='Upper Bound')
plt.plot(input_sizes, linear_model(input_sizes, *linear_coeffs), '--', linestyle='--', color='purple', label='Lower Bound')
plt.xlabel('n')
plt.ylabel('Time (seconds)')
plt.title('Execution Time vs n')
plt.legend()
plt.grid(True)
plt.show()
```
